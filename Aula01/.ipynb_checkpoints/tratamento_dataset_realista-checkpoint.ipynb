```python
#importar bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
```

```python
df_realista=pd.read_csv('dataset_feedback_realista.csv') #ler csv
```

```python
df_realista.head(40) #mostrar primeiras 40 linhas
```

```python
#informações geriais, é possivel ver os tipos de dados, total de colunas, quantos valores não nulos 
df_realista.info()
```

```python
df_realista.describe() #Informações estatísticas
```

```python
df_realista.describe(include='all') #descrição com mais detalhes
```

```python
#utlizando métodos pandas para achar as medidas estatísticas
mediana_idade = df_realista['idade'].median()
mediana_nota = df_realista['nota'].median()
media_renda = df_realista['renda'].mean()

print(f'Mediana das idades {mediana_idade}')
print(f'Mediana das Notas {mediana_nota}')
print(f'valor medio das rendas {media_renda:.4f}')
```

```python
#Calculo IQRe outliers para renda, idade e nota

Q1_renda = df_realista['renda'].quantile(0.25)
Q3_renda = df_realista['renda'].quantile(0.75)
IQR_renda = Q3_renda - Q1_renda
limite_inferior_renda = Q1_renda - 1.5 * IQR_renda
limite_superior_renda = Q3_renda + 1.5 * IQR_renda

outliers_renda=df_realista[(df_realista['renda'] < limite_inferior_renda) | (df_realista['renda'] > limite_superior_renda)]

#---------------------------------------------------------

Q1_idade = df_realista['idade'].quantile(0.25)
Q3_idade = df_realista['idade'].quantile(0.75)
IQR_idade = Q3_idade - Q1_idade
limite_inferior_idade = Q1_idade - 1.5 * IQR_idade
limite_superior_idade = Q3_idade + 1.5 * IQR_idade

outliers_idade=df_realista[(df_realista['idade'] < limite_inferior_idade) | (df_realista['idade'] > limite_superior_idade)]

#---------------------------------------------------------
Q1_nota = df_realista['nota'].quantile(0.25)
Q3_nota = df_realista['nota'].quantile(0.75)
IQR_nota = Q3_nota - Q1_nota
limite_inferior_nota = Q1_nota - 1.5 * IQR_nota
limite_superior_nota = Q3_nota + 1.5 * IQR_nota

outliers_nota=df_realista[(df_realista['nota'] < limite_inferior_nota) | (df_realista['nota'] > limite_superior_nota)]

print(f'\n\n\t Calculo de IQR e Outliers \n\n')
print(f'\n\t IQR - Campo Renda \n')
print(f'IQR ----------- {IQR_renda:.2f}')
print(f'Quartis -- Q1: {Q1_renda:-8.2f}   -> Q3: {Q3_renda:-8.2f}')
print(f'Limites -- Inf: {limite_inferior_renda:-5.2f}  -> Sup: {limite_superior_renda:-8.2f}')
print(f'\nPossíveis outliers em "renda":\n{outliers_renda}\n')


print(f'\n\t IQR - Campo Idade \n')
print(f'IQR ----------- {IQR_idade:.2f}')
print(f'Quartis -- Q1: {Q1_idade:-6.2f}   -> Q3: {Q3_idade:-8.2f}')
print(f'Limites -- Inf: {limite_inferior_idade:-5.2f}  -> Sup: {limite_superior_idade:-7.2f}')
print(f'\nPossíveis outliers em "idade":\n{outliers_idade}\n')


print(f'\n\t IQR - Campo Nota \n')
print(f'IQR ----------- {IQR_nota:.2f}')
print(f'Quartis -- Q1: {Q1_nota:-5.2f}   -> Q3: {Q3_nota:-8.2f}')
print(f'Limites -- Inf: {limite_inferior_nota:-5.2f}  -> Sup: {limite_superior_nota:-7.2f}')
print(f'\nPossíveis outliers em "nota": \n{outliers_nota}\n')


```

```python
# Observação: os valores negativos nos limites inferiores não indicam erro.
# Eles são resultado direto da fórmula do IQR: Q1 - 1.5 * IQR.
# Mesmo que o limite inferior seja negativo, isso só afetará a identificação de outliers
# se existirem valores reais menores que esse limite — o que não acontece, por exemplo,
# em colunas como idade, renda ou nota (que não podem ser negativas).
```

```python
#contar valores nulos 
df_realista.isnull().sum()
```

```python
#importante tratar valores nulos
# para os valores numericos podemos substituir os valores pelas medianas e media encontradas anteriormente  
df_realista['idade']=df_realista['idade'].fillna(mediana_idade)
df_realista['nota']=df_realista['nota'].fillna(mediana_nota)
df_realista['renda']=df_realista['renda'].fillna(media_renda)


#para os valores NÃO numericos podemos colocar "Não informado" nas celulas nulas
df_realista['cidade']=df_realista['cidade'].fillna('Não informado')
df_realista['feedback']=df_realista['feedback'].fillna('Não informado')
```

```python
#visualizar ultimas 20 linhas 
df_realista.tail(20)
```

```python
df_realista.isnull().sum() #contando numeros nulos após o tratamento
```

```python
# Codificação das colunas categóricas 'cidade', 'categoria' e 'feedback' com get.dummies .
# transforma cada categoria distinta em uma nova coluna binária,
# facilitando a análise e uso desses dados em modelos estatísticos e de machine learning.
# O parâmetro drop_first=False garante que todas as categorias sejam mantidas (sem exclusão da primeira)
#dtype=int  # Garante que os valores serão 0 ou 1 (int) em vez de True/False (bool)
df_encoded = pd.get_dummies(df_realista, columns=['cidade','categoria','feedback'], drop_first=False, dtype=int)
```

```python
df_encoded.head(10)
```

```python
df_encoded.info()
```

```python
# Definimos uma lista com os nomes das colunas numéricas que queremos padronizar (normalizar):
numeric_cols = ['idade', 'renda', 'nota']

# Criamos um objeto StandardScaler, que vai aplicar a padronização: transformar os dados para que
# tenham média 0 e desvio padrão 1.
scaler = StandardScaler()

# Aplicamos a padronização nas colunas numéricas do DataFrame 'df_encoded'. O método fit_transform
# calcula a média e o desvio padrão de cada coluna e transforma os valores para a escala padronizada.
# Substituímos as colunas originais pelas colunas padronizadas.
df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])
```
