```python
#continuação do df_final mas com base do papagaio

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from joblib import dump, load
```

```python
df_final=pd.read_csv('dataset_tratado_sudeste_realista_outliers-fase1.csv')
```

```python
df_final.head()
```

```python
# x = todas as colunas exceto "categoria"
# x = df_final.drop('categoria', axis=1)

x = df_final[['idade', 'renda', 'nota']]
# y = a coluna que queremos prever 
y = df_final['categoria']
```

```python
x.head()
```

```python
y.head()
```

```python
print(f"\n\nFeatures \n\n{x.head()} \n")
```

```python
print(f"\n\nTarget \n\n{y.head()} \n")
```

```python
X_train, X_test, y_train, y_test=train_test_split(
    x, y,
    test_size=0.2,     # 20% dos dados serão usados para teste
    random_state=42,   # garantir reprodutibilidade
    stratify=y         # manter proporção das classes em treino e teste
)
```

```python
print(f"Dados de Treino - Features \n\n{X_train.head()} \n\n\n")
print(f"Dados de Treino - Target \n\n{y_train.head()} \n\n\n")

print(f"Dados de Treino - Features \n\n{X_train.head()} \n\n\n")
print(f"Dados de Treino - Target \n\n{y_test.head()} \n\n\n")
```

```python
modelo=LogisticRegression(max_iter=1000)
```

```python
#treinando modelo
modelo.fit(X_train, y_train)
```

```python
y_pred=modelo.predict(X_test)
```

```python
acuracia=accuracy_score(y_test, y_pred)
print(f'Acurácia do modelo: {acuracia:.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
